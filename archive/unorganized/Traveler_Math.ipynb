# Traveler Math

The BFS which carried the the Traveler lasted from 2020 to 2046. The signal was transmitted across multiple frequencies and had a presitant bandwidth of 4 TB/s.



```python
from math import *

import numpy as np
import matplotlib.pyplot as plt

years_to_seconds = 365 * 24 * 60 * 60

years = 2046 - 2020
seconds = years * years_to_seconds

signal_bandwidth = 4 # TB/s

traveler_size = seconds * signal_bandwidth # TB
traveler_size = traveler_size // 2
traveler_size = traveler_size // 1000 // 1000 # TB -> EB
traveler_size_bytes = traveler_size * 1e18 # EB -> bytes

print(f"Size of the Traveler: {traveler_size} EB")
```

The Traveler inputs a binary stream and outputs another binary stream and a
cybernetic stream to update its weights (all integrated, somehow, into one
neural network, unlike human multi-organ AI models).

Each unit of the binary stream, split by time and size, is assigned to a vector
with $n$ dimensions (the max throughput of the Traveler in bytes). No additional
embedding takes place; time and where the data was split are not included in the
vector.

This vector is then passed through $m$ hidden layers, each with weights and
biases, all with $n$ dimensions. The output binary stream is the same size as
the input (although 50% is reserved for the cybernetic loop).

The size of the model is then given by:

$$ T = m(n^2 + n) $$

Since $T$ is the size of the model in bytes, which is known:

$$ m = \frac{T}{n^2+n} $$




```python
n = np.linspace(1e6,1e10,1000)
m = traveler_size_bytes / (n + n**2)

trav_n = 2**28 # dim (bytes)
trav_m = traveler_size_bytes // (trav_n + trav_n**2)

plt.loglog(n, m,label="Possible models")
plt.scatter(trav_n, trav_m,label=f"({trav_n}, {trav_m})")
plt.legend()
```
